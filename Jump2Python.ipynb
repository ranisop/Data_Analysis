#### 2. 문자열 포매팅(Formatting)

* 숫자 바로 대입 → %d

```python
'I eat %d apples.' %3
```

* 문자열 바로 대입 → %s

```python
'I eat %s apples.' %'five'
```

* 2개 이상의 값 넣기    
###### 2개 이상의 값을 넣으려면 마지막 % 다음 괄호 안에 콤마(,)로 구분하여 각각의 값을 넣어주면 된다.

```python
number = 10
day = 'three'
'I ate %d apples. so I was sick for %s days.' %(number, day)
```

#### 3. 문자열 포맷 코드

코드 | 설명    
:-----: | :-----
%s | 문자열(String)
%c | 문자 1개(Character)
%d | 정수(Integer)
%f | 부동소수(Floating-point)
%o | 8진수
%x | 16진수
%% | Literal % (문자 % 자체)

* `%s` 는 자동으로 % 뒤에 있는 값을 문자열로 바꾼다.

```python
'I have %s apples' % 3
```

```python
'rate is %s' % 3.234
```

* 포매팅 연산자 %d와 %를 같이 쓸 때는 `%%`를 쓴다.

* `ValueError: incomplete format`    
###### %를 나타내려면 반드시 %%로 써야한다.

```python
'Error is %d%.' %98
```

```python
'Error is %d%%.' %98
```

#### 4. 포맷 코드와 숫자 함께 사용하기

* 정렬과 공백    
`%10s` : 전체 길이가 10개인 문자열 공간에서 오른쪽 정렬

```python
'%10s' %'hi'
```

```python
'%-10sjane' %'hi'
```

* 소수점 표현하기

```python
'%0.4f' %3.42143234
```

```python
'%10.4f' %3.42143234
```

#### 5. Format 함수를 사용한 포매팅

```python
# 숫자 바로 대입하기
'I eat {0} apples.'.format(3)
```

```python
# 문자열 바로 대입하기
'I eat {0} appels.'.format('five')
```

```python
# 숫자 값을 가진 변수로 대입하기
number = 3
'I eat {0} apples.'.format(number)
```

```python
# 2개 이상의 값 넣기
number = 10
day = 'three'
'I ate {0} apples. so I was sick for {1} days.'.format(number, day)
```

```python
# 이름으로 넣기
'I ate {number} apples. so I was sick for {day} days.'.format(number=10, day='three')
```

```python
# 인덱스와 이름을 혼용해서 넣기
'I ate {0} apples. so I was sick for {day} days.'.format(10, day='three')
```

* 정렬    
`:<10` : 치환되는 문자열을 왼쪽으로 정렬하고 문자열의 총 자릿수를 10으로 맞춘다.    
`:<` 대신 `:>` 를 사용하면 오른쪽 정렬    
`:^` 은 가운데 정렬

```python
# 왼쪽 정렬
'{0:<10}'.format('hi')
```

```python
# 오른쪽 정렬
'{0:>10}'.format('hi')
```

```python
# 가운데 정렬
'{0:^10}'.format('hi')
```

* 공백 채우기    
정렬 문자 `<, >, ^` 바로 앞에 지정한 문자를 넣으면 해당 값으로 공백이 채워진다.

```python
'{0:=^10}'.format('hi')
```

```python
'{0:!^10}'.format('hi')
```

* 소수점 표현하기

```python
y = 3.42134234
'{0:0.4f}'.format(y)
```

```python
'{0:10.4f}'.format(y)
```

* `{` 또는 `}` 문자 표현하기

```python
'{{ and }}'.format()
```

#### 6. f 문자열 포매팅
문자열 앞에 `f` 접두사를 붙이면 f 문자열 포매팅 기능을 사용할 수 있다.

```python
name = '홍길동'
age = 30
f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
```

```python
age = 30
f'나는 내년이면 {age+1}살이 된다.'
```

```python
# 딕셔너리에서 사용
d = {'name':'홍길동', 'age':30}
f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
```

```python
# 왼쪽 정렬
f'{"hi":<10}'
```

```python
# 오른쪽 정렬
f'{"hi":>10}'
```

```python
# 가운데 정렬
f'{"hi":^10}'
```

```python
# 가운데 정렬 + '=' 공백 채우기
f'{"hi":=^10}'
```

```python
# 왼쪽 정렬 + '!' 공백 채우기
f'{"hi":!<10}'
```
