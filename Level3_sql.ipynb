```python
import sqlite3
import pandas as pd
```

### Load Dataset

```python
# 'e-commerce.sqlite' 파일이 있으면 로드하고 없으면 새로 만들라는 뜻이기 때문에
# 로드하는 파일명이 틀려도 에러가 안남! 
# 경로/파일명 정확하게 하기
connection = sqlite3.connect('e-commerce.sqlite')
connection
```

### SELECT

```python
# orders 전체 다 가져오기
query = "select * from 'orders'"  

order = pd.read_sql(query, connection, index_col = 'id')
# 인덱스 지정 : index_col = ' '

order
```

```python
# 컬럼 지정해서 가져오기
query = "select id, date, price from 'orders'"  

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
# 행 갯수 일부분 가져오기
query = "select * from 'orders' limit 3"  

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
# order by : default asc(빠른순), desc(늦은순)
query = "select * from 'orders' order by date desc"  

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# sql 보기 좋게 정리!
query = """
    select *
        from 'orders'
        order by price desc
        limit 3
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

### WHERE (조건)

```python
# 조건 <, >
query = """
    select *
        from 'orders'
        where price > 500
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 조건 = (일치)
query = """
    select *
        from 'orders'
        where state = 'confirmed'
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 조건 <> (불일치)
query = """
    select *
        from 'orders'
        where state <> 'confirmed'
"""
# != 는 가끔 안먹히는 곳이 있음
# <> 는 sql에서 공통적으로 다 쓰는 거니까 이거로 쓰기

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 조건 between 
query = """
    select *
        from 'orders'
        where price between 400 and 900
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 조건 in
query = """
    select *
        from 'orders'
        where address in ('Seoul', 'Incheon')
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 조건 and, or
query = """
    select *
        from 'orders'
        where price >= 500 and state = 'confirmed'
        order by price desc
        limit 3
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 조건 like
query = """
    select *
        from 'orders'
        where address like '%a%'
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

### 여러 개의 데이터를 하나로 합치기

```python
query = """
    select *
        from 'orders'
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
query = """
    select *
        from 'users'
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
query = """
    select *
        from 'products'
"""

order = pd.read_sql(query, connection, index_col='id')
order
```

```python
# 공통된 key로 합치기(select *)
query = """
    select *
        from 'orders'
    left join 'users' on 'orders'.user_id = 'users'.id
"""

order = pd.read_sql(query, connection)
order
```

> `join`할 때 select * 해서 합치면 겹치는 컬럼이 있을 수 있어서    
> select * 하지 않고 `컬럼을 지정`해서 합쳐주는 것이 좋다!

```python
# 공통된 key로 합치기(컬럼 지정)
query = """
    select 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender
        from 'orders'
    left join 'users' on 'orders'.user_id = 'users'.id
"""

order = pd.read_sql(query, connection)
order
```
