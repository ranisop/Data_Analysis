```python
import pandas as pd
```

### 1. Load Dataset

```python
url = 'http://bit.ly/dsa-01-order'

# 데이터 읽기 전부터 index 알 때는 index_col
# order = pd.read_csv(url , index_col='id')  

# 데이터 읽고 난 다음에 index 설정할 때는 set_index
order = pd.read_csv(url)

# order에 다시 넣지 않고 inplace 옵션으로 설정할수도 있음
# order.set_index('id'. inplace=True)
order = order.set_index('id')
```

### 3. Pivot table
* 대용량 데이터에 최적화    
* 통계 연산에서 기본적으로 NaN은 제외
* aggfunc : default는 평균(mean)

```python
pd.pivot_table(order, index='state', values='price')
```

```python
pd.pivot_table(order, index='address', values='price')
```

```python
pd.pivot_table(order, index='state', values='price', aggfunc='sum')
```

```python
pd.pivot_table(order, index='state', values='price', aggfunc=['mean', 'sum'])
```

```python
import numpy as np

pd.pivot_table(order, index='state', values='price', aggfunc=[np.mean, np.sum])
```

```python
pd.pivot_table(order, index=['date', 'state'], values='price')
```

### 4. Group by
* pivot table과 거의 비슷
* 통계 연산을 해서 눈으로 확인하고 싶으면 pivot table(예쁘게 보임)
* 통계 연산을 한 결과를 가지고 다른 연산을 하거나 파일/DB에 저장 등을 하기 위해서는 group by
* group by가 속도가 더 빠르기 때문에 결과 출력이 목적이 아니라면 그냥 group by 쓰는게 좋다.

```python
order['price'].mean()
```

```python
order
```

```python
# state가 confirmed/canceled 일때의 mean
order.groupby('state')['price'].mean()
```

```python
order.groupby(['date', 'state'])['price'].sum()
```

```python
%timeit pd.pivot_table(order, index=['date', 'state'], values='price')
```

```python
%timeit order.groupby(['date', 'state'])['price'].sum()
```

```python
# Method Chaining

table = pd.pivot_table(order, index=['date', 'state'], values='price')

# table
# table['price']
# table.loc['confirmed']
table.sort_values(by='price', ascending=False).head(3)
```

### 5. apply
* 컬럼명이랑 함수의 인자는 동일할 필요가 없다.

```python
def is_vip(price):
    if price >= 500:
        return 'VIP'
    else:
        return 'Non-VIP'

order['price'].apply(is_vip)
```

```python
# 컬럼 하나가 아니라 여러개 쓰고 싶으면 전체에 apply 적용하기
def is_vip(row):
    if row['state'] == 'confirmed' and row['price'] >= 500:
        return 'VIP'
    else:
        return 'Non-VIP'

order.apply(is_vip, axis='columns')
```

### 6-2. merge

```python
# order + user
# 키가 다를 때 : left_on, right_on
pd.merge(order, user, left_on='user_id', right_on='id')
```

```python
# 키가 같을 때 : on
user['user_id'] = user.index

pd.merge(order, user, on='user_id')
```

```python
# 키를 지정하지 않으면 pandas가 알아서 같은 이름을 찾아서 합침
# 이름만 같을 뿐 다른 데이터일수도 있기 때문에 문제가 생길수도 있다.
pd.merge(order, user)
```

```python
# pd.merge(order, product)  # 키 지정 안하면 price로 합쳐짐

# price가 두개라서 x,y로 들어감 -> 컬럼 이름 바꿔줘야 함
pd.merge(order, product, left_on='product_id', right_on='id')
```

```python
# inner : 교집합(default)
pd.merge(order, user, on='user_id', how='inner')
```

```python
# outer : 합집합
pd.merge(order, user, on='user_id', how='outer')
```

```python
# left : 왼쪽을 살리고, 오른쪽을 버림
pd.merge(order, user, on='user_id', how='left')
```

```python
# right : 오른쪽을 살리고, 왼쪽을 버림
pd.merge(order, user, on='user_id', how='right')
```

> ⭐️ 메인 데이터를 왼쪽에 두고 how='left'로 다 합치는 걸 추천
